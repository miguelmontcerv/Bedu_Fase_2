# Speed dating: ¿Cómo llegar a una segunda cita?

## Planteamiento del Problema

Los solteros en México representan el 31.4 por ciento de las personas mayores de 15 años, según datos revelados este año por el INEGI. Para conocer gente y así terminar con su soltería este 31.4 por ciento recurre a diferentes métodos, entre ellos invitar a salir a sus amig@s y aplicaciones como Tinder, Grindr, etc. Existe una forma de conocer gente que busca pareja un tanto conocida que es en la que se basa este análisis y es el Speed Dating.

### ¿Qué es Speed Dating?

La expresión inglesa speed dating se traduce en español como cita rápida y consiste en organizar una cita múltiple de solteros, con un número idéntico de mujeres y hombres. El evento consiste en tener una cita de varios minutos de duración, normalmente entre 8 y 12 , con cada uno de los solteros del sexo opuesto que acudan.

La dinámica es simple. Se enumera a los presentes en parejas para comenzar con una cita de cinco minutos. El timbre de una campana indica el cambio de los asistentes masculinos a otra mesa siguiendo la numeración de las féminas que los esperan; todo termina hasta dar una vuelta completa. En una tarjeta (Score Card) se apunta el número de la persona a la que le agradaría conocer. Si las personas hacen ‘click’ y se eligen mutuamente, la empresa se encargará de proporcionarles el contacto (correo o número telefónico opcional), para que inicie la interacción a ritmo propio.


El origen del Speed Date se remonta a 1998, en Los Ángeles, California. El rabino judío Yaacov Deyo congregó a hombres y mujeres para propiciar casamientos. Una matraca que se usa en la festividad del Purim avisaba el término de los 10 minutos que duraba esa interacción. Tras popularizarse, el modelo fue copiado, incluso por programas de televisión, contra los cuales no se pudo hacer nada para patentar el sistema y poco a poco se convirtió en una práctica que comparten distintos países del mundo.

El speed dating es una manera de conocer gente y encontrar pareja cada vez más extendida y, aunque se pueden organizar citas informales, cada vez existen más empresas que se dedican a organizar este tipo de eventos.
